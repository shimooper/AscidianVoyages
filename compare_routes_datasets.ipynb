```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
SHIPS_ROUTES_DATASET = r"C:\Users\yairs\OneDrive\Documents\University\Master\Ships\datasets from Doron\ships routes\ships routes (ports only) with locations of ports.xlsx"
SHIPS_ROUTES_DATASET_EXTENDED = r"C:\Users\yairs\OneDrive\Documents\University\Master\Ships\datasets from Doron\ships routes\ships routes (ports only) with locations of ports - extended.csv"
SHIPS_ROUTES_NO_LOCATIONS_DATASET = r"C:\Users\yairs\OneDrive\Documents\University\Master\Ships\datasets from Doron\ships routes\ship routes (ports only).xlsx"
```

```python
def analyze_ships_routes_df(routes_df):
    routes_df['count'] = 1
    ships_routes_length = routes_df.groupby(['Ship']).count()['count']
    print(ships_routes_length)
    print(f'Total num of ships: {len(ships_routes_length)}')
    print(f'Longest route: {ships_routes_length.max()}')
    print(f'Shortest route: {ships_routes_length.min()}')
    sns.histplot(ships_routes_length, binwidth=1, kde=True)
```

```python
routes_df = pd.read_excel(SHIPS_ROUTES_DATASET)
analyze_ships_routes_df(routes_df)
```

```python
routes_extended_df = pd.read_csv(SHIPS_ROUTES_DATASET_EXTENDED)
analyze_ships_routes_df(routes_extended_df)
```

```python
routes_no_locations_df = pd.read_excel(SHIPS_ROUTES_NO_LOCATIONS_DATASET)
analyze_ships_routes_df(routes_no_locations_df)
```
